```python
import pandas as pd
import numpy as np
```

# pd.Series

```python
a = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'],dtype='float')
print(a)
```

```python
a = pd.Series({'A': 1, 'B': 2,'C': 3, 'D': 4, 'E': 5},dtype='float', name = 'use dic to create a series')
print(a)
```

 # access to the ___element___

```python
a = pd.Series({'A': 1, 'B': 2,'C': 3, 'D': 4, 'E': 5},dtype='float', name = 'use dic to create a series')
print(pd.Series(a,index = ['C','L','E']))
```

# __PROPERTY__ of series

```python
a = pd.Series({'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5}, dtype='float', name='use dic to create a series')
print(a.index)
print(a.values)
print(a.shape, a.ndim, a.size)

print(a.loc['B'])
print(a.iloc[1])
print(a.loc['B':'D'])  # the last index is concluded
print(a.iloc[1:3])

print(a.at['B'])
print(a.iat[1])
```

# s[key] is an appropriate way to access to corresponding value
# but _s[index]_ is inappropriate

```python
print(a['A'])
```

# filtering

```python
a = pd.Series(np.arange(10), index=map(lambda x: x.upper(), ['a','b','c','d','e','f','g','h','i','j']))
print(a)
print(a[a>5])
```
