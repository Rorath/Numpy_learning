# expand_dims(nparray, axis)

```python
a = np.arange(10).reshape(2, 5)
print(a, a.shape, a.ndim)
print('-----------------------')
for _ in range(a.ndim + 1):
    b = np.expand_dims(a, axis=_)
    print(b, b.shape)
```
